# (A) - Categorical EDA

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
ax = df.groupby("person_home_ownership").size().plot(kind="bar",color=["green","red","yellow","orange"])
for container in ax.containers:
    ax.bar_label(container)
plt.title("count of home ownership type")
plt.subplot(2,2,2)
df.groupby("person_home_ownership").size().plot(kind="pie",autopct="%0.2f")
plt.title("Percentage of home ownership type")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 14))
plt.subplot(3,3,1)
sns.histplot(df["person_age"], bins=10, color="red")  # color should be a string, not a list
plt.title("Age Distribution Among Age Categories")
plt.xlabel("Age")
plt.ylabel("Frequency")



plt.subplot(3,3,2)
sns.histplot(df["person_income"], bins=10, color="purple")  # color should be a string, not a list
plt.title("Person Income Distribution")
plt.ylabel("Person Income")
plt.xlabel("Frequency")





plt.subplot(3,3,3)
sns.histplot(df["person_emp_length"], bins=10, color="blue")  # color should be a string, not a list
plt.title("person Emp Length Distibution")
plt.xlabel("person Emp Length Distibution")
plt.ylabel("Frequency")



plt.subplot(3,3,4)
sns.histplot(df["loan_amnt"], bins=10, color="orange")  # color should be a string, not a list
plt.title("Loan Amount Distribution")
plt.xlabel("Loan Amount")
plt.ylabel("Frequency")






plt.subplot(3,3,5)
sns.histplot(df["loan_int_rate"], bins=10, color="orange")  # color should be a string, not a list
plt.title("Loan Interest Rate Distribution")
plt.xlabel("Loan Interest Rate")
plt.ylabel("Frequency")



plt.subplot(3,3,6)
sns.histplot(df["loan_percent_income"], bins=10, color="orange")  # color should be a string, not a list
plt.title("Loan Percent Income")
plt.xlabel("Loan Percent")
plt.ylabel("Frequency")




plt.subplot(3,3,7)
sns.histplot(df["cb_person_cred_hist_length"], bins=10, color="orange")  # color should be a string, not a list
plt.title("Person Credit History length Distibution")
plt.xlabel("Person Credit History length")
plt.ylabel("Frequency")\

plt.tight_layout()
plt.show()
```
